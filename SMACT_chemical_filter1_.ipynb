```python
import smact
from smact import Element, neutral_ratios
from smact.screening import pauling_test
import multiprocessing
from pymatgen.core import Composition

def setup_binary_oxide_space():
    """Set up the chemical space of transition metal + oxygen"""
    transition_metals = ["Ti", "V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn"]
    tm_elements = [Element(symbol) for symbol in transition_metals]
    oxygen = Element("O")
    return tm_elements, oxygen

def binary_oxide_filter(metal: Element):
    """Filter binary oxides for charge neutrality and electronegativity"""
    compounds = []
    o_state = -2  # Oxygen oxidation state

    for metal_ox in metal.oxidation_states:
        # Check charge-neutral ratios
        try:
            cn_exists, cn_ratios = neutral_ratios([metal_ox, o_state], threshold=8)
        except Exception:
            continue

        if cn_exists:
            eneg_ok = pauling_test(
                [metal_ox, o_state],
                [metal.pauling_eneg, 3.44]  # 3.44 is electronegativity of oxygen
            )
            if eneg_ok:
                formula = [metal.symbol, "O"]
                compounds.append((formula, cn_ratios[0]))  # get only first viable ratio

    return compounds

def generate_binary_oxides():
    """Main function to generate viable binary oxides"""
    metals, oxygen = setup_binary_oxide_space()

    with multiprocessing.Pool() as pool:
        binary_results = pool.map(binary_oxide_filter, metals)

    # Format results
    binary_formulas = []
    for result in binary_results:
        for comp in result:
            formula = "".join(f"{el}{amt if amt != 1 else ''}" for el, amt in zip(comp[0], comp[1]))
            reduced = Composition(formula).reduced_formula
            binary_formulas.append(reduced)

    return sorted(set(binary_formulas))

# Run the binary oxide screening
viable_binary_oxides = generate_binary_oxides()

# Display the results
print("\n Viable Binary Oxide Candidates for Photocatalysis or Semiconducting Applications:")
for ox in viable_binary_oxides:
    print(f" - {ox}")
```

```python
import smact
from smact import Element, neutral_ratios
from smact.screening import pauling_test
import multiprocessing
from pymatgen.core import Composition

def setup_chalcogenide_space():
    """Define ternary chalcogenide chemical space for solar cells"""
    group_11 = ["Cu", "Ag"]
    group_13 = ["In", "Ga"]
    chalcogens = ["S", "Se"]

    metal_1 = [Element(m) for m in group_11]
    metal_2 = [Element(m) for m in group_13]
    chalc = [Element(c) for c in chalcogens]

    return metal_1, metal_2, chalc

def ternary_chalcogenide_filter(elements):
    """Apply charge neutrality and electronegativity filters to ternary chalcogenides"""
    m1, m2, ch = elements
    compounds = []

    for ox_1 in m1.oxidation_states:
        for ox_2 in m2.oxidation_states:
            for ox_ch in ch.oxidation_states:
                ox_states = [ox_1, ox_2, ox_ch]

                try:
                    cn_exists, cn_ratios = neutral_ratios(ox_states, threshold=8)
                except Exception:
                    continue

                if cn_exists:
                    # Check electronegativity match
                    eneg_ok = pauling_test(
                        ox_states, [m1.pauling_eneg, m2.pauling_eneg, ch.pauling_eneg]
                    )
                    if eneg_ok:
                        formula = [m1.symbol, m2.symbol, ch.symbol]
                        stoich = cn_ratios[0]
                        compounds.append((formula, stoich))
    return compounds

def generate_ternary_chalcogenides():
    """Main routine to generate ternary chalcogenides"""
    m1_els, m2_els, ch_els = setup_chalcogenide_space()
    combinations = [(m1, m2, ch) for m1 in m1_els for m2 in m2_els for ch in ch_els]

    with multiprocessing.Pool() as pool:
        results = pool.map(ternary_chalcogenide_filter, combinations)

    # Format formulas
    formulas = []
    for group in results:
        for (elements, stoichs) in group:
            formula = "".join(f"{el}{amt if amt > 1 else ''}" for el, amt in zip(elements, stoichs))
            reduced = Composition(formula).reduced_formula
            formulas.append(reduced)

    return sorted(set(formulas))

# Run it
ternary_chalcogenides = generate_ternary_chalcogenides()

# Display
print("\n🔆 Viable Ternary Chalcogenides for Solar Applications:")
for formula in ternary_chalcogenides:
    print(f" - {formula}")
```

```python
import smact
from smact import Element
from smact.screening import pauling_test
import multiprocessing
from pymatgen.core import Composition
import pandas as pd

def setup_double_perovskite_space():
    """Define chemical space for A2BB'O6 double perovskites"""
    a_site = ["Ba", "Sr", "Ca"]                  # Large A-site cations
    b_site = ["Fe", "Mn", "Ni"]                  # Transition metals
    b_prime_site = ["Mo", "W", "Re"]             # d⁰ or partially filled TM
    oxygen = Element("O")

    a_els = [Element(el) for el in a_site]
    b_els = [Element(el) for el in b_site]
    bp_els = [Element(el) for el in b_prime_site]

    return a_els, b_els, bp_els, oxygen

def double_perovskite_filter(elements):
    """Filter double perovskites based on charge neutrality and electronegativity"""
    a, b, bp, o = elements
    viable = []

    for a_ox in a.oxidation_states:
        for b_ox in b.oxidation_states:
            for bp_ox in bp.oxidation_states:
                total_charge = 2*a_ox + b_ox + bp_ox + 6*(-2)
                if total_charge == 0:
                    eneg_ok = pauling_test(
                        [a_ox, b_ox, bp_ox, -2],
                        [a.pauling_eneg, b.pauling_eneg, bp.pauling_eneg, 3.44]
                    )
                    if eneg_ok:
                        elements_list = [a.symbol, b.symbol, bp.symbol, "O"]
                        stoich = [2, 1, 1, 6]
                        viable.append((elements_list, stoich))
    return viable

def generate_double_perovskites():
    """Run full filter and return unique formulas"""
    a_els, b_els, bp_els, oxygen = setup_double_perovskite_space()
    combinations = [(a, b, bp, oxygen) for a in a_els for b in b_els for bp in bp_els]

    with multiprocessing.Pool() as pool:
        raw_results = pool.map(double_perovskite_filter, combinations)

    flat_results = [item for sublist in raw_results for item in sublist]
    formulas = []

    for elements, stoichs in flat_results:
        formula = "".join(f"{el}{amt if amt > 1 else ''}" for el, amt in zip(elements, stoichs))
        reduced = Composition(formula).reduced_formula
        formulas.append(reduced)

    unique_formulas = sorted(set(formulas))
    return unique_formulas

# Run generator
perovskite_candidates = generate_double_perovskites()

# Display results
print("\n✅ First 10 Unique Double Perovskite Candidates:")
for i, formula in enumerate(perovskite_candidates[:10], 1):
    print(f" {i:2d}. {formula}")

print(f"\n🔢 Total unique double perovskite formulas: {len(perovskite_candidates)}")
```

```python
from smact.utils.oxidation import ICSD24OxStatesFilter
from smact import Element
from smact.screening import smact_filter
from pymatgen.core import Composition
from itertools import combinations

# === Step 1: Define battery-relevant elements ===
element_symbols = ["Na", "Fe", "O"]
elements = [Element(sym) for sym in element_symbols]

def generate_valid_compositions(el_list, oxidation_states_set="icsd24", maxstoichiometrythreshold=8):
    """Return unique reduced Composition objects from smact_filter."""
    combos = smact_filter(
        el_list,
        threshold=maxstoichiometrythreshold,
        oxidation_states_set=oxidation_states_set,
        species_unique=True
    )
    comps = set()
    for symbols, ox_states, ratios in combos:
        comp = Composition({sym: amt for sym, amt in zip(symbols, ratios)}).reduced_composition
        comps.add(comp)
    return comps

def generate_filtered_compositions(oxidation_set_path=None, maxstoichiometrythreshold=8):
    """Generate ternary + binary SMACT-filtered compositions for battery materials."""
    ox_set = oxidation_set_path or "icsd24"

    # Ternary: Na–Fe–O
    ternary = generate_valid_compositions(elements, ox_set, maxstoichiometrythreshold)

    # Binary: All combinations of Na, Fe, O
    binary = set()
    for pair in combinations(elements, 2):
        binary |= generate_valid_compositions(list(pair), ox_set, maxstoichiometrythreshold)

    all_comps = ternary | binary
    return all_comps, ternary, binary

# === Step 2: Compare different commonality thresholds ===
ox_filter = ICSD24OxStatesFilter()
commonality_thresholds = [1, 2, 5, 10]
results_comparison = {}

print("🔬 Comparing SMACT results across commonality thresholds...\n")
for threshold in commonality_thresholds:
    custom_file = f"oxidation_states_battery_commonality_{threshold}.txt"

    # Write oxidation state file with custom commonality threshold
    ox_filter.write(
        custom_file,
        consensus=1,
        include_zero=False,
        commonality=threshold,
        comment=f"Na–Fe–O system | Commonality ≥ {threshold}"
    )

    # Run SMACT filtering
    try:
        all_comps, ternary_comps, binary_comps = generate_filtered_compositions(custom_file)
        results_comparison[threshold] = {
            'total': len(all_comps),
            'ternary': len(ternary_comps),
            'binary': len(binary_comps)
        }
    except Exception as e:
        print(f" Error with threshold {threshold}: {e}")
        results_comparison[threshold] = {'total': 0, 'ternary': 0, 'binary': 0}

# === Step 3: Display results ===
print("\n Comparison of SMACT Filtering by Commonality Threshold (Na–Fe–O)")
print(f"{'Threshold':>10} | {'Total':>8} | {'Ternary':>8} | {'Binary':>8}")
print("-" * 50)
for threshold, counts in results_comparison.items():
    print(f"{threshold:>10} | {counts['total']:>8} | {counts['ternary']:>8} | {counts['binary']:>8}")

print("\n Interpretation: Higher commonality thresholds = stricter filtering = fewer compositions (more reliable)")
```
